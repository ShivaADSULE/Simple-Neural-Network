```python
import numpy as np # its required for multiplication
```

```python
weights = np.array([0.7,0.2,-0.5]) # taking the random weights
alpha = 0.1 # setting the model weight correction step size that means effect of every iteration
```

```python
streetlight = np.array([
    [0,0,1],
    [0,1,1],
    [0,0,1],
    [1,1,1],
    [0,1,1],
    [1,0,1]
]) #input data
walk_vs_stop = np.array([0,1,0,1,1,0]) #output data
```

```python
for iterations in range(40): # taking 40 iterations
    error_for_all = 0        # to get total error for all inputs
    for row in range(len(walk_vs_stop)):    # taking every input
        ip = streetlight[row]   # ip
        goal_pred = walk_vs_stop[row] #op
        prediction = ip.dot(weights) # prediction
        error = (prediction - goal_pred)**2 # finding error
        error_for_all +=error # add_to total error
        #to modify the weights
        delta = prediction - goal_pred # find the current differnce in the prediction
        """
        multiply ip*delta is done due to there the 
        1.stoping
        2.negative reversal
        3.scaling
        """
        weights = weights - (alpha*(ip * delta)) # get the weights

        print(" PREDICTION :",prediction)
    print(" weights :",weights)
    print(" errors :",error_for_all)
    
```
